 ### 1 Creating and Manipulating Arrays
 ## First import numpy and matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1.1  Create two 2D arrays representing coordinates x, y on the cartesian plan
## Both should cover the range (-2, 2) and have 100 points in each direction

```python
x=np.linspace(-2,2,100)
y=np.linspace(-2,2,100)

xx,yy=np.meshgrid(x,y)
print(xx)
```

### 1.2 Visualize each 2D array using pcolormesh

```python
plt.pcolormesh(x,y,xx,cmap='magma')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
```

```python
plt.pcolormesh(x,y,yy,cmap='magma')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
```

### From your cartesian coordinates, create polar coordinates r  an $\phi$


```python
r = np.sqrt(xx**2+yy**2)
phi=np.arctan2(yy,xx)
```

### 1.4. Visualize r and $\phi$ on the 2D x/y using pcolormeshmesh

```python
plt.pcolormesh(x,y,r,cmap='magma')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
```

```python
plt.pcolormesh(x,y,phi,cmap='twilight')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
```

### 1.5 Caclulate the quanity f=cos2(4r)+sin2(4phi)

```python
f=(np.cos(4*r)**2)+(np.sin(4*phi)**2)

plt.pcolormesh(x,y,f,cmap='magma')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
```

### 1.6 Plot the mean of f with respect to the x axis as a function of y

```python
plt.plot(x,f.mean(axis=0))
plt.xlabel('Y')
plt.ylabel('Mean f')
```

### 1.7  Plot the mean of f with respect to the y axis as a function of x

```python
plt.plot(y,f.mean(axis=1))
plt.ylabel('Mean f')
plt.xlabel('X')
```

### 1.8 Plot the mean of f with respect to $\Phi$ as a function of r 


```python
r_range=np.linspace(0,2*np.sqrt(2),1000) # 2 sqrt 2 is max value of r when x = 2 and y = 2
phi_range=np.linspace(-np.pi,np.pi,1000)
storemean=np.zeros([1000,1000]) 
for i in range(len(r_range)):
    for j in range(len(phi_range)):
        f=(np.cos(4*r_range[i])**2)+(np.sin(4*phi_range[j])**2)
        storemean[i,j]=np.mean(f)

plt.pcolormesh(r_range,phi_range,storemean)
plt.xlabel('r')
plt.ylabel('$\phi$')
    
```

### 2.4 Compute the mean and standard deviation of each of T, S and P at each depth in level.

```python



T_mean=[]
T_std=[]
P_mean=[]
P_std=[]
S_mean=[]
S_std=[]

for i in range(len(level)):
    tt=np.mean(T[i,:])
    T_mean.append(tt)
    ts=np.std(T[i,:])
    T_std.append(ts)
    pt=np.mean(P[i,:])
    P_mean.append(pt)
    ps=np.std(P[i,:])
    P_std.append(ps)
    st=np.mean(S[i,:])
    S_mean.append(st)
    ss=np.std(S[i,:])
    S_std.append(ss)


plt.errorbar(T_mean,level,xerr=T_std)
plt.xlabel('Temperature')
plt.ylabel('Level')
plt.gca().invert_yaxis()
```

```python
plt.errorbar(S_mean,level,xerr=S_std)
plt.xlabel('Salinity')
plt.ylabel('Level')
plt.gca().invert_yaxis()
```

```python
plt.errorbar(P_mean,level,xerr=P_std)
plt.xlabel('Pressure')
plt.ylabel('Level')
plt.gca().invert_yaxis()
```

### 2.6 Account For Missing Data

```python



T_mean=[]
T_std=[]
P_mean=[]
P_std=[]
S_mean=[]
S_std=[]

for i in range(len(level)):
    tt=np.nanmean(T[i,:])
    T_mean.append(tt)
    ts=np.nanstd(T[i,:])
    T_std.append(ts)
    pt=np.nanmean(P[i,:])
    P_mean.append(pt)
    ps=np.nanstd(P[i,:])
    P_std.append(ps)
    st=np.nanmean(S[i,:])
    S_mean.append(st)
    ss=np.nanstd(S[i,:])
    S_std.append(ss)


plt.errorbar(T_mean,level,xerr=T_std)
plt.xlabel('Temperature')
plt.ylabel('Level')
plt.gca().invert_yaxis()
```

```python
plt.errorbar(S_mean,level,xerr=S_std)
plt.xlabel('Salinity')
plt.ylabel('Level')
plt.gca().invert_yaxis()
```

```python
plt.errorbar(P_mean,level,xerr=P_std)
plt.xlabel('Pressure')
plt.ylabel('Level')
plt.gca().invert_yaxis()
```
